```python
# Bai 4.20
import pandas as pd

# Đọc dữ liệu
euro12 = pd.read_csv('euro12.csv')

# 1. In giá trị cột Goals
print(euro12['Goals'])

# 2. Có bao nhiêu đội tham gia?
so_doi = euro12.shape[0]
print(f"Số đội tham gia: {so_doi}")

# 3. In thông tin của Euro2012
print(euro12.info())

# 4. Tạo dataframe mới tên là 'ky_luat' (discipline)
ky_luat = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# 5. Sắp xếp giảm dần theo 'Red Cards' sau đó đến 'Yellow Cards'
ky_luat_sap_xep = ky_luat.sort_values(['Red Cards', 'Yellow Cards'], ascending=False)

# 6. a) Tính trung bình Yellow Cards
tb_the_vang = ky_luat['Yellow Cards'].mean()

# b) Lọc các đội ghi hơn 6 bàn thắng
doi_tren_6_ban = euro12[euro12['Goals'] > 6]

# 7. In các đội bắt đầu bằng chữ 'G'
doi_chu_G = euro12[euro12['Team'].str.startswith('G')]

# 8. In 7 cột đầu của euro12
print(euro12.iloc[:, :7])

# 9. In tất cả các cột, trừ 3 cột cuối
print(euro12.iloc[:, :-3])

# 10. In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])

# 11. In cột Team và Shooting Accuracy của England, Italy, Russia
print(euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']])
```

```python
#Bài 4.21 
import pandas as pd

# 1. Đọc dữ liệu với index_col là cột đầu tiên
ruou_bia = pd.read_csv('drinks.csv', index_col=0)

# In kiểu dữ liệu, kích thước, tên cột và 5 dòng đầu/cuối
print(type(ruou_bia))
print(ruou_bia.shape)
print(ruou_bia.columns)
print(ruou_bia.head())
print(ruou_bia.tail())

# 2. Số lượng bia tiêu thụ trung bình ở mỗi châu lục
bia_tb_chau_luc = ruou_bia.groupby('continent')['beer_servings'].mean()

# 3. Thông tin thống kê tổng quát (describe) rượu vang theo châu lục
vang_thong_ke = ruou_bia.groupby('continent')['wine_servings'].describe()

# 4. Số lượng trung bình (mean) các loại bia và rượu theo mỗi châu lục
tb_tat_ca = ruou_bia.groupby('continent').mean()

# 5. Giá trị trung vị (median) cho các loại bia và rượu theo mỗi châu lục
trung_vi_tat_ca = ruou_bia.groupby('continent').median()

# 6. Rượu mạnh (spirit_servings): trung bình, lớn nhất, nhỏ nhất theo châu lục
ruou_manh_stats = ruou_bia.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])

# 7. Sắp xếp tăng dần theo số lượng bia tiêu thụ
sap_xep_bia = ruou_bia.sort_values('beer_servings')

# 5 quốc gia tiêu thụ bia NHIỀU nhất
top_5_nhieu = ruou_bia.sort_values('beer_servings', ascending=False).head(5)

# 5 quốc gia tiêu thụ bia ÍT nhất
top_5_it = ruou_bia.sort_values('beer_servings', ascending=True).head(5)
```
