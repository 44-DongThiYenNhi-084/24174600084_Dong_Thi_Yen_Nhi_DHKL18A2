```python
#Bai 4.4 
import pandas as pd

chuoi1 = pd.Series([1, 2, 3, 4, 5])
chuoi2 = pd.Series([4, 5, 6, 7, 8])

# 1. Xóa các mục của chuoi1 có mặt trong chuoi2
ket_qua_1 = chuoi1[~chuoi1.isin(chuoi2)]

# 2. Lấy các phần tử chỉ xuất hiện ở 1 trong 2 (không nằm chung)
giao_nhau = pd.Series(list(set(chuoi1) & set(chuoi2)))
ket_qua_2 = pd.concat([chuoi1, chuoi2])
ket_qua_2 = ket_qua_2[~ket_qua_2.isin(giao_nhau)]
```

```python
#Bai 4.5 
import numpy as np
import pandas as pd

du_lieu = pd.Series(np.random.normal(10, 5, 25))

# Tính các giá trị thống kê
min_val = du_lieu.min()            # Giá trị tối thiểu
p25 = np.percentile(du_lieu, 25)   # Phân vị thứ 25
trung_vi = du_lieu.median()        # Trung vị
p75 = np.percentile(du_lieu, 75)   # Phân vị thứ 75
max_val = du_lieu.max()            # Giá trị tối đa
```

```python
# Bai 4.7 
import numpy as np
import pandas as pd

trang_thai = np.random.RandomState(100)
chuoi_so = pd.Series(trang_thai.randint(1, 5, [12]))

# Tìm 2 giá trị xuất hiện nhiều nhất
top2 = chuoi_so.value_counts().index[:2]

# Thay các giá trị còn lại bằng 'Other'
chuoi_so[~chuoi_so.isin(top2)] = 'Other'
```
