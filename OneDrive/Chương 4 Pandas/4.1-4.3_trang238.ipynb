```python
# Bài 4.1: 
import pandas as pd
import numpy as np

# Dữ liệu đầu vào
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

# Cách làm đơn giản nhất: dùng pd.Series()
ser1 = pd.Series(mylist)
ser2 = pd.Series(myarr)
ser3 = pd.Series(mydict)

# Xem kết quả thử 1 cái
print(ser3.head()) # head() để xem 5 dòng đầu cho gọn
```

```python
#Bai 4.2
import pandas as pd

# Dữ liệu từ bài trước
ser = pd.Series(dict(zip(list('abcdefg'), range(7))))

# Bước 1: Chuyển sang DataFrame bằng to_frame()
# Bước 2: Dùng reset_index() để đưa cột chỉ mục vào trong bảng
df = ser.to_frame().reset_index()

# Đặt lại tên cột cho đẹp (tùy chọn)
df.columns = ['Chu_cai', 'So_tuong_ung']

print(df.head())
```

```python
#Bai 4.3 
import pandas as pd
import numpy as np

ser1 = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))

# Cách 1: Dùng hàm concat (ghép) theo trục dọc hoặc ngang
# axis=1 nghĩa là ghép theo chiều ngang để thành 2 cột
df = pd.concat([ser1, ser2], axis=1)

# Cách 2: Tạo DataFrame bằng cách bỏ vào dictionary (Dễ hiểu nhất)
df = pd.DataFrame({'Cot_1': ser1, 'Cot_2': ser2})

print(df.head())
```
