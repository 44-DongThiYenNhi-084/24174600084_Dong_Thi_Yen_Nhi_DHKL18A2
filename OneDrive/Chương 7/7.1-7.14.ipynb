```python
#Bai 7.1 
import tkinter as tk
from tkinter import messagebox

def kiem_tra_dien_du(event):
    # Nếu cả 3 ô đều có chữ thì hiện nút gửi
    if o_ten.get() and o_id.get() and o_mat_khau.get():
        nut_gui.pack(pady=10)
    else:
        nut_gui.pack_forget()

def thong_bao_gui():
    messagebox.showinfo("Thông báo", "Đã gửi thông tin thành công!")

cua_so = tk.Tk()
cua_so.title("Đăng ký thông tin")
cua_so.geometry("300x250")

tk.Label(cua_so, text="Student Name:").pack()
o_ten = tk.Entry(cua_so)
o_ten.pack()
o_ten.bind("<KeyRelease>", kiem_tra_dien_du)

tk.Label(cua_so, text="Student ID:").pack()
o_id = tk.Entry(cua_so)
o_id.pack()
o_id.bind("<KeyRelease>", kiem_tra_dien_du)

tk.Label(cua_so, text="Password:").pack()
o_mat_khau = tk.Entry(cua_so, show="*")
o_mat_khau.pack()
o_mat_khau.bind("<KeyRelease>", kiem_tra_dien_du)

nut_gui = tk.Button(cua_so, text="Submit", command=thong_bao_gui)
# Ban đầu ẩn nút gửi
nut_gui.pack_forget()

cua_so.mainloop()
```

```python
#BAI 7.6 
import tkinter as tk

def dao_nguoc_chuoi():
    chuoi_vao = o_nhap.get()
    # Cách đảo ngược thủ công không dùng hàm [::-1]
    chuoi_nguoc = ""
    for ky_tu in chuoi_vao:
        chuoi_nguoc = ky_tu + chuoi_nguoc
    nhan_ket_qua.config(text=chuoi_nguoc)

cua_so = tk.Tk()
cua_so.title("Đảo ngược từ")

tk.Label(cua_so, text="Enter a word:").grid(row=0, column=0, padx=10, pady=10)
o_nhap = tk.Entry(cua_so)
o_nhap.grid(row=0, column=1)

nhan_ket_qua = tk.Label(cua_so, text="", fg="blue")
nhan_ket_qua.grid(row=1, column=1)

tk.Button(cua_so, text="Validate", command=dao_nguoc_chuoi).grid(row=2, column=1, pady=10)

cua_so.mainloop()
```

```python
#Bai 7.7 
import tkinter as tk

def tinh_tong():
    try:
        n = int(o_n.get())
        tong = sum(range(1, n + 1))
        bieu_thuc = " + ".join(map(str, range(1, n + 1)))
        nhan_ket_qua.config(text=f"The sum is {bieu_thuc} = {tong}")
    except:
        nhan_ket_qua.config(text="Vui lòng nhập số nguyên!")

cua_so = tk.Tk()
cua_so.title("Tính tổng")

tk.Label(cua_so, text="Enter value of integer N:").pack(pady=5)
o_n = tk.Entry(cua_so)
o_n.pack(pady=5)

nhan_ket_qua = tk.Label(cua_so, text="", wraplength=250)
nhan_ket_qua.pack(pady=10)

tk.Button(cua_so, text="Validate", command=tinh_tong).pack(pady=5)

cua_so.mainloop()
```

```python
#Bai 7.8 
import tkinter as tk

def tim_uoc_so():
    try:
        n = int(o_n.get())
        danh_sach_uoc = [str(i) for i in range(1, n + 1) if n % i == 0]
        nhan_ket_qua.config(text=f"The divisors of N: {' '.join(danh_sach_uoc)}")
    except:
        nhan_ket_qua.config(text="Lỗi nhập liệu!")

cua_so = tk.Tk()
cua_so.title("Tìm ước số")

tk.Label(cua_so, text="Enter the value of N:").pack(pady=5)
o_n = tk.Entry(cua_so)
o_n.pack(pady=5)

nhan_ket_qua = tk.Label(cua_so, text="")
nhan_ket_qua.pack(pady=10)

tk.Button(cua_so, text="Validate", command=tim_uoc_so).pack(pady=5)

cua_so.mainloop()
```

```python
#Bai 7.9 
import tkinter as tk
from tkinter import ttk
import math

def thuc_hien():
    try:
        m = int(o_m.get())
        n = int(o_n.get())
        lua_chon = combo_chon.get()
        
        if lua_chon == "GCD":
            kq = math.gcd(m, n)
            nhan_kq.config(text=f"GCD({m}, {n}) = {kq}")
        else:
            kq = abs(m * n) // math.gcd(m, n)
            nhan_kq.config(text=f"LCM({m}, {n}) = {kq}")
    except:
        nhan_kq.config(text="Nhập m và n hợp lệ!")

cua_so = tk.Tk()
cua_so.title("GCD & LCM")

tk.Label(cua_so, text="Enter value of m:").grid(row=0, column=0)
o_m = tk.Entry(cua_so)
o_m.grid(row=0, column=1)

tk.Label(cua_so, text="Enter value of n:").grid(row=1, column=0)
o_n = tk.Entry(cua_so)
o_n.grid(row=1, column=1)

tk.Label(cua_so, text="Choose function:").grid(row=2, column=0)
combo_chon = ttk.Combobox(cua_so, values=["GCD", "LCM"])
combo_chon.current(0)
combo_chon.grid(row=2, column=1)

nhan_kq = tk.Label(cua_so, text="", fg="red")
nhan_kq.grid(row=3, columnspan=2)

tk.Button(cua_so, text="Calculate", command=thuc_hien).grid(row=4, columnspan=2)

cua_so.mainloop()
```

```python
#Bai 7.10 
import tkinter as tk
from tkinter import messagebox

def kiem_tra_tuoi():
    ten = o_ten.get()
    try:
        tuoi = int(o_tuoi.get())
        if tuoi >= 18:
            msg = f"Xin chào {ten}. Bạn đã trưởng thành!"
        else:
            msg = f"Xin chào {ten}. Bạn còn nhỏ tuổi!"
        messagebox.showinfo("Thông báo", msg)
    except:
        messagebox.showerror("Lỗi", "Vui lòng nhập tuổi là số!")

cua_so = tk.Tk()
cua_so.title("Kiểm tra độ tuổi")

tk.Label(cua_so, text="Nhập tên:").pack()
o_ten = tk.Entry(cua_so)
o_ten.pack()

tk.Label(cua_so, text="Nhập tuổi:").pack()
o_tuoi = tk.Entry(cua_so)
o_tuoi.pack()

tk.Button(cua_so, text="Kiểm tra", command=kiem_tra_tuoi).pack(pady=10)

cua_so.mainloop()
```

```python
#Bai 7.12 
import tkinter as tk
from tkinter import messagebox

def tinh_bmi():
    try:
        can_nang = float(o_can_nang.get())
        chieu_cao = float(o_chieu_cao.get())
        so_bmi = can_nang / (chieu_cao ** 2)
        
        # Phân loại theo chuẩn
        if so_bmi < 18.5:
            ket_luan = "Gầy"
        elif 18.5 <= so_bmi < 24.9:
            ket_luan = "Bình thường"
        elif 25 <= so_bmi < 29.9:
            ket_luan = "Thừa cân"
        else:
            ket_luan = "Béo phì"
            
        nhan_ket_qua.config(text=f"BMI: {so_bmi:.2f}")
        nhan_phan_loai.config(text=f"Kết luận: {ket_luan}")
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ!")

cua_so = tk.Tk()
cua_so.title("BMI Calculator")
cua_so.geometry("300x250")

tk.Label(cua_so, text="Cân nặng (kg):").pack(pady=5)
o_can_nang = tk.Entry(cua_so)
o_can_nang.pack()

tk.Label(cua_so, text="Chiều cao (m):").pack(pady=5)
o_chieu_cao = tk.Entry(cua_so)
o_chieu_cao.pack()

tk.Button(cua_so, text="Tính BMI", command=tinh_bmi).pack(pady=10)

nhan_ket_qua = tk.Label(cua_so, text="BMI: --", font=("Arial", 10, "bold"))
nhan_ket_qua.pack()

nhan_phan_loai = tk.Label(cua_so, text="Kết luận: --", fg="blue")
nhan_phan_loai.pack()

cua_so.mainloop()
```

```python
#Bai 7.13 
import tkinter as tk
from tkinter import messagebox

def hien_thong_bao(chuong, bai):
    messagebox.showinfo(f"Chương {chuong}", f"Giải bài tập {bai}")

cua_so = tk.Tk()
cua_so.title("Hệ thống bài tập Python")
cua_so.geometry("400x300")

thanh_menu = tk.Menu(cua_so)

# Tạo các menu con từ Chương 1 đến Chương 4
for i in range(1, 5):
    menu_chuong = tk.Menu(thanh_menu, tearoff=0)
    for j in range(1, 10):
        # Sử dụng lambda để truyền tham số vào hàm
        menu_chuong.add_command(label=f"{i}.{j}", command=lambda c=i, b=j: hien_thong_bao(c, b))
    thanh_menu.add_cascade(label=f"Chương {i}", menu=menu_chuong)

cua_so.config(menu=thanh_menu)
tk.Label(cua_so, text="Chọn bài tập từ Menu phía trên", font=("Arial", 12)).pack(expand=True)

cua_so.mainloop()
```

```python
#Bai 7.14 
import tkinter as tk

def tinh_tien_taxi():
    try:
        km = float(o_km.get())
        phut_cho = float(o_phut_cho.get())
        loai_xe = bien_loai_xe.get()
        tong_tien = 0

        # Tính tiền chờ (miễn phí 5 phút đầu)
        tien_cho = 0
        if phut_cho > 5:
            tien_cho = (phut_cho - 5) * 750

        if loai_xe == 4:  # Xe 4 chỗ
            gia_mo_cua = 12000 # cho 0.8km đầu
            if km <= 0.8:
                tong_tien = gia_mo_cua
            elif km <= 30:
                tong_tien = gia_mo_cua + (km - 0.8) * 15300
            else:
                tong_tien = gia_mo_cua + (29.2 * 15300) + (km - 30) * 12100
        else:  # Xe 7 chỗ
            gia_mo_cua = 12000 # cho 0.8km đầu
            if km <= 0.8:
                tong_tien = gia_mo_cua
            elif km <= 30:
                tong_tien = gia_mo_cua + (km - 0.8) * 16100
            else:
                tong_tien = gia_mo_cua + (29.2 * 16100) + (km - 30) * 13800
        
        nhan_kq.config(text=f"Tổng cước: {tong_tien + tien_cho:,.0f} VNĐ")
    except ValueError:
        nhan_kq.config(text="Vui lòng nhập số km và phút chờ!")

cua_so = tk.Tk()
cua_so.title("Tính cước Taxi")

tk.Label(cua_so, text="Số km di chuyển:").pack()
o_km = tk.Entry(cua_so)
o_km.pack()

tk.Label(cua_so, text="Số phút chờ:").pack()
o_phut_cho = tk.Entry(cua_so)
o_phut_cho.insert(0, "0")
o_phut_cho.pack()

bien_loai_xe = tk.IntVar(value=4)
tk.Radiobutton(cua_so, text="Xe 4 chỗ", variable=bien_loai_xe, value=4).pack()
tk.Radiobutton(cua_so, text="Xe 7 chỗ", variable=bien_loai_xe, value=7).pack()

tk.Button(cua_so, text="Tính tiền", command=tinh_tien_taxi, bg="orange").pack(pady=10)
nhan_kq = tk.Label(cua_so, text="Tổng cước: 0 VNĐ", font=("Arial", 11, "bold"))
nhan_kq.pack()

cua_so.mainloop()
```
