```python
import xml.dom.minidom
def main(): 
    doc= xml.dom.minidom.parse("sample.xml")

    print(doc.nodeName) # in nút gốc
    print(doc.documentElement.tagName) # in phần tử gốc 
    company= doc.getElementsByTagName("staff") #lấy các thẻ staff
    for st in company: 
        name= st.getElementsByTagName("name")[0].firstChild.data # lấy nội dung trong thẻ tên 
        salary= st.getElementsByTagName("salary")[0].firstChild.data # lấy nội dung trong thẻ lương
        print(name) #in 
        print(salary) # in

# tạo nhân viên mới
    new_staff = doc.createElement("staff") #tạo phần tử mới
    name_new_staff = doc.createElement("name")
    name_new_staff.appendChild(doc.createTextNode("Galasa"))
    salary_new_staff = doc.createElement("salary")
    salary_new_staff.appendChild(doc.createTextNode("4.6LPA"))


# thêm tên , lương vào thẻ nhân viên mới (new_staff)
    new_staff.appendChild(name_new_staff)
    new_staff.appendChild(salary_new_staff)

# thêm vào new_staff gốc 
    doc.documentElement.appendChild(new_staff)

#thêm nhân viên mới vào danh sách và in 
    staff_list = doc.getElementsByTagName("staff")
    print("Danh sách nhân viên sau khi đã thêm: ")
    for st in staff_list: 
        name= st.getElementsByTagName("name")[0].firstChild.data # lấy nội dung trong thẻ tên 
        salary= st.getElementsByTagName("salary")[0].firstChild.data # lấy nội dung trong thẻ lương
        print(name) #in 
        print(salary) # in

if __name__=="__main__": 
    main();

# Đề bài 2.4 phân tích file sample.xml bai tap 3










```
