```python
import xml.dom.minidom 
def main():
    doc= xml.dom.minidom.parse("sample.xml") #doc file xml 
    print(doc.nodeName) # in ra nút gốc 
    print(doc.firstChild.tagName) #in ra phần tử gốc  

    company = doc.getElementsByTagName("staff") # lấy các nút Staff
    for st in company:  # duyệt đến từng phần tử staff trong company 
        name= st.getElementsByTagName("name")[0].firstChild.data #lấy nội dung tên trong mỗi phần tử staff
        print(name)


if __name__=="__main__":
    main();

#Đề bài 2.5 dùng phương thức getElementsByTagName() để lấy các phần tử sau đó in ra tên các phần tử 

```
