```python
# vd 3.4.2 
import numpy as np
import timeit # Import thư viện timeit để đo thời gian thực thi của các đoạn code.

# Tính tổng của mảng sử dụng Numpy
np_setup = "import numpy as np; np_array = np.arange(1, 1000001)"
np_code  = "np_sum = np.sum(np_array)"
np_time  = timeit.timeit(stmt=np_code, setup=np_setup, number=10)

# Tính tổng của danh sách sử dụng List
list_setup = "list_array = list(range(1, 1000001))"
list_code = "list_sum = sum(list_array)"
list_time = timeit.timeit(stmt=list_code, setup=list_setup, number=10)

print("Thời gian tính toán của Numpy:", np_time)
print("Thời gian tính toán của List:", list_time)

print("Numpy tính nhanh hơn List ở lượt chạy là ~ ", int(list_time/np_time), "lần!!")
```

```python
#vd 3.4.3.1
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

#Phép Cộng:Sử dụng hàm np.add() hoặc toán tử '+'.
Tong = np.add(arr1,arr2)
#hoặc tong = arr1+arr2

#Phép Trừ: Sử dụng hàm np.subtract() hoặc toán tử '-'.
Hieu = np.subtract(arr1,arr2)
#hoặc Tru = arr1-arr2

#Phép Nhân: Sử dụng hàm np.multiply() hoặc toán tử '*'
Tich = np.multiply(arr1,arr2)
#Hoặc arr1*arr2

#Phép Chia: Sử dụng hàm np.divide() hoặc toán tử '/'
Thuong = np.divide(arr1,arr2)

print("Tổng của arr1 + arr2 là:",Tong)
print("Hiệu của arr1 - arr2 là:",Hieu)
print("Tích của arr1 * arr2 là:",Tich)
print("Thương của arr1/arr2 là:",Thuong)
```

```python
#vd 3.4.3.3 
import numpy as np

# Tạo mảng dữ liệu với kích thước lớn (1000 phần tử)
data = np.random.randint(100, 200, size=1000)

# Tạo một giá trị trung bình (số thực)
mean_value = np.array([data.mean()])

# Tính tỷ lệ phần trăm tăng trưởng bằng cách áp dụng broadcasting
percentage_growth= (data - mean_value) / mean_value * 100

# In ra 10 phần tử đầu tiên của mảng dữ liệu và tỷ lệ phần trăm tăng trưởng
print("Dữ liệu gốc:", data[:10])
print("Tỷ lệ phần trăm tăng trưởng:", percentage_growth[:10])
```
