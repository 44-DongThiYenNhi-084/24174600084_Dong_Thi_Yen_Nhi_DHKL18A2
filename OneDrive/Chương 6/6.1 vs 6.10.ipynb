```python
# Bai 6.1  
import threading
import time

# Hàm mà các luồng sẽ thực thi
def cong_viec_cua_luong():
    ten_luong = threading.current_thread().name
    print(f"Luồng '{ten_luong}' đang chạy...")
    time.sleep(1) # Giả lập công việc tốn thời gian
    print(f"Luồng '{ten_luong}' đã hoàn thành.")

# Danh sách chứa các luồng
danh_sach_luong = []

# Tạo ra 5 luồng khác nhau
for i in range(5):
    ten = f"Luong_So_{i+1}"
    luong_moi = threading.Thread(target=cong_viec_cua_luong, name=ten)
    danh_sach_luong.append(luong_moi)
    luong_moi.start() # Bắt đầu chạy luồng

# Chờ tất cả các luồng kết thúc trước khi kết thúc chương trình chính
for luong in danh_sach_luong:
    luong.join()

print("Tất cả các luồng đã thực hiện xong.")
```

```python
#Bai 6.10 
import threading
import random

# Hàm tìm giá trị lớn nhất trong một đoạn danh sách
def tim_max_trong_doan(danh_sach_con, ket_qua_tam, chi_so):
    gia_tri_lon_nhat = max(danh_sach_con)
    ket_qua_tam[chi_so] = gia_tri_lon_nhat
    print(f"Luồng {chi_so} tìm thấy max trong đoạn là: {gia_tri_lon_nhat}")

# 1. Tạo danh sách ngẫu nhiên từ 0-100 với n phần tử (ví dụ 1000 phần tử)
n = 1000
danh_sach_tong = [random.randint(0, 100) for _ in range(n)]

# 2. Chia làm 4 luồng để thực hiện
so_luong = 4
kich_thuoc_con = n // so_luong
danh_sach_luong = []
ket_qua_tam = [0] * so_luong

for i in range(so_luong):
    # Chia nhỏ danh sách cho từng luồng
    bat_dau = i * kich_thuoc_con
    ket_thuc = (i + 1) * kich_thuoc_con if i != so_luong - 1 else n
    doan_con = danh_sach_tong[bat_dau:ket_thuc]
    
    # Khởi tạo và chạy luồng
    luong = threading.Thread(target=tim_max_trong_doan, args=(doan_con, ket_qua_tam, i))
    danh_sach_luong.append(luong)
    luong.start()

# 3. Chờ các luồng làm xong
for luong in danh_sach_luong:
    luong.join()

# 4. Tìm giá trị lớn nhất cuối cùng từ các kết quả tạm của các luồng
max_cuoi_cung = max(ket_qua_tam)
print("-" * 30)
print(f"Giá trị lớn nhất trong toàn bộ danh sách là: {max_cuoi_cung}")
```
